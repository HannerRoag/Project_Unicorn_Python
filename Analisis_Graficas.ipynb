Hay que agregar detalles adicionales al gráfico

```python
plt.figure(figsize = (9,4))
plt.plot(df.fecha[df.serie == 'Total'], df.poblacion[df.serie == 'Total']/1000000)
plt.xlabel('Year')
plt.ylabel('Poblation(Million)')
plt.title('Spanish-Total Poblation',pad =20)
plt.show();
```

Cuando tenemos una visualización, la podemos guardar en función

```python
plt.figure(figsize = (9,4))
plt.plot(df.fecha[df.serie == 'Mujeres'], df.poblacion[df.serie == 'Mujeres']/1000000)
plt.xlabel('Year')
plt.ylabel('Poblation(Million)')
plt.title('Spanish-Fem Poblation',pad =20)
plt.show();
```

```python
plt.figure(figsize = (9,4))
plt.plot(df.fecha[df.serie == 'Hombres'], df.poblacion[df.serie == 'Hombres']/1000000)
plt.xlabel('Year')
plt.ylabel('Poblation(Million)')
plt.title('Spanish-Men Poblation',pad =20)
plt.show();
```

Podemos plotear las series seguidas.

```python
df.serie.unique()
```

```python
plt.figure(figsize = (9,4))
for i in df.serie.unique():
    plt.plot(df.fecha[df.serie == i], df.poblacion[df.serie == i]/1000000)

plt.legend()
plt.xlabel('Year')
plt.ylabel('Poblation(Million)')
plt.title('Spanish Poblation (1971 - 2022)', pad =20)
plt.show();
```

Cuando esté construyendo cada una de las lineas, queremos que el i sea etiquetada

```python
plt.figure(figsize = (9,4))
for i in df.serie.unique():
    plt.plot(df.fecha[df.serie == i], df.poblacion[df.serie == i]/1000000,label = i)

plt.legend()
plt.xlabel('Year')
plt.ylabel('Poblation(Million)')
plt.title('Spanish Poblation (1971 - 2022)', pad =20)
plt.show();
```

Serie total es mas grande que las otras series, cuando hacemos la interpretación, vemos que las lineas de hombres y mujeres están muy pegadas, podemos modificar las proporciones. Para entrar en detalle, dejamos las series separadas del total, y poder comparar entre ellas. 

```python
plt.figure(figsize = (9,4))
for i in ['Hombres','Mujeres']:
    plt.plot(df.fecha[df.serie == i], df.poblacion[df.serie == i]/1000000,label = i)

plt.legend()
plt.xlabel('Year')
plt.ylabel('Poblation(Million)')
plt.title('Spanish Poblation (1971 - 2022)', pad =20)
plt.show();
```

Encontramos lo mismo que en datos numéricos, donde el crecimiento en los hombres y en las mujeres, no tenían la misma pendiente.
